### Pie Charts

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Import dataset
pokemon = pd.read_csv('data/pokemon.csv')
```

#### Example 1. Plot a simple Pie chart

```python
# Use the same pokemon dataset
sorted_counts = pokemon['generation_id'].value_counts()

plt.pie(sorted_counts, 
        labels = sorted_counts.index, 
        startangle = 90, 
        counterclock = False);
```

#### TODO: Using another list of strings to display labels on each wedge

```python
labels = [f'gen{gen}' for gen in sorted_counts.index]
labels
```

```python
plt.pie(sorted_counts, 
        labels = labels, 
        startangle = 90, 
        counterclock = False);
```

### Donut Plot

#### Example 2. Plot a simple Donut plot

```python
plt.pie(sorted_counts, 
        labels = labels, 
        startangle = 90,
        counterclock = False, 
        wedgeprops = {'width' : 0.4});
```
